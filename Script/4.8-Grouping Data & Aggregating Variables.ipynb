# Aggregating Data with transform()

## create a loyality flag using loc()

```python
# group by 'user_id', tranform fun. on order_number and create a new column as max_order

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Deriving Columns with loc()

## to create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value.

```python
# to create a new column as 'loyality_flag'

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'Loyalty_flag_loc'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order']<= 40), 'Loyalty_flag_loc'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<= 10, 'Loyalty_flag_loc'] = 'New customer'
```

```python
# check the frequency 

ords_prods_merge['Loyalty_flag_loc'].value_counts(dropna = False)
```

## Q.n.7

```python
# group by 'user_id', tranform fun. on 'days_since_prior_order' and create a new column as order_median

ords_prods_merge['order_median']= ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
# check the head

ords_prods_merge.head()
```

```python
# Create an order frequency flag according to the median in the “days_since_prior_order” column

ords_prods_merge.loc[ords_prods_merge['order_median'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_median'] > 10) & (ords_prods_merge['order_median'] <= 20),'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_median'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check the frequency

ords_prods_merge['order_frequency_flag'].value_counts(dropna= False)
```
