# 6.Combine your customer data with the rest of your prepared Instacart data.

```python
# Import prepared Instacart data

ords_prods_merge = pd.read_pickle(os.path.join(PATH, 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
ords_prods_merge.head()
```

```python
# check the datatype of 'user_id'in ords_prods_merge data

data_type = ords_prods_merge['user_id'].dtype
print(data_type)
```

```python
df_customer.head()
```

```python
# change the 'user_id' datatype in df_customer data

df_customer['user_id'] = df_customer['user_id'].astype(float)
```

```python
# check the datatype of 'user_id'in df_customer data

data_type = df_customer['user_id'].dtype
print(data_type)
```

```python
# merge the df_customer with ords_prods_merge dataframe

df_instacart = df_customer.merge(ords_prods_merge ,on = 'user_id')
```
